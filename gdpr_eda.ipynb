```python
# modul import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
df = pd.read_csv('enforcement_tracker_full_data.csv')
```

```python
display(df.head())
display(df.tail())
```

```python
# Changing the column names
df.rename(columns={'Date of Decision': 'Date_of_Decision'}, inplace=True)
df.rename(columns={'Fine [€]': 'Fine_EUR'}, inplace=True) 
df.rename(columns={'Controller/Processor': 'Controller_Processor'}, inplace=True)
df.rename(columns={'Quoted Art.': 'Quoted_Art'}, inplace=True)
df.rename(columns={'Direct URL': 'Direct_URL'}, inplace=True)
```
